# Question 4
Pick a suitable spacing for $t = 0:0.01:8$ and $v = -8:0.01:8$ and then use the subplot to plot the following functions,

>$z = e^{-0.5t}cos(20t - 6) \quad \text{for } 0 \le t \le 8$ \\
>$u = 6log10(v^{2} +20)  \quad \text{for } -8 \le v \le 8$


```
t = arange(0,8,0.01)
v = arange(-8,8,0.01)

z = (exp(-0.5*t))*(cos(20*t + 6))
u = 6*log10((v**2) + 20)
subplot(2,1,1)
plot(t,z)
title('z(t)')
xlabel('t')
ylabel('z')

subplot(2,1,2)
plot(v,u)
title('u(v)')
xlabel('v')
ylabel('u')
```

# Question 7
Plot the following functions using axes that will produce a straight-line
plot. The power function is $y = 2x^{-0.5}$, and the exponential function is $y = 10^{1-x}$. 
>$0.01 \le x \le 3, loglog, semilogy$

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.01,3,0.0001)
y1 = 2*(x**-0.5)
y2 = 10**(1-x)

plt.subplot(1,2,1)
plt.loglog(x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(1,2,2)
plt.semilogy(x,y2)
plt.xlabel('x')
plt.ylabel('y')
```

# Question 8
Plot the function $y = 8x^{3} for -1 \le x \le 1$ with a tick spacing of 0.25
on the 𝑥 axis and 2 on the y axis.

```
x = np.arange(-1,1,0.01)
y = 8*(x**3)
plt.plot(x,y)
plt.title('y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.annotate('y=8x^3',(-0.5,0.25))
plt.xticks(np.arange(min(x), max(x)+0.25, 0.25));
plt.yticks(np.arange(min(y), max(y)+2, 2));
```

# Question 9
Obtain the plot of the following implicit function, known as the
Ampersand curve. Use the axis equal command.
>$(y^{2} - x^{2})(x-1)(2x - 3) = 2(x^{2} + y^{2} + 2x^{2})$

axis equal sets the aspect ratio so that the data units are the same in every
direction.

```
# Matplotlib does not have implicit plotting capabilities
# Sympy.plot_implicit needs to be used
from sympy import plot_implicit,symbols, Eq, And,plot
x,y= symbols('x y')
eqn = Eq((y**2 - x**2)*(x - 1)*(2*x - 3) - 2*((x**2 + y**2 - 2*x)**2))
plot_implicit(eqn,title='Ampersand Curve',x_var=(x, -0.25, 1.5),y_var=(y, -1.5, 1.5), adaptive=False)
```

# Question 10
Use plot3 and fplot3 to plot the 3-D line plot described by $x = sin(t), y = cos(t), z = ln(t) for 0 < t < 30$.


```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

t = np.arange(0.01,30,0.01)
x = np.sin(t)
y = np.cos(t)
z = np.log(t)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x,y,z)
plt.xticks(np.arange(-1,1.25,0.5))
plt.yticks(np.arange(-1,1.25,0.5))
```
