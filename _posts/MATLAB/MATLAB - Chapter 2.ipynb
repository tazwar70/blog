# Question 11
Given vectors
$x = 
\begin{bmatrix}
6\\
7\\
3
\end{bmatrix}
$
and
$y = 
\begin{bmatrix}
2 & 4 & 7
\end{bmatrix}
$

Find w=x*y and z=y*x

```
x = np.array([[6],[7],[3]])
y = np.array([2,4,7])
# for scalar multiplication, use *
w = x*y
# matmul does matrix multiplication
z = np.matmul(y,x)
print(f'w = \n{w}\n\nz = \n{z}')
```

# Question 13
Solve the following set of equations

>$14x - 2y = 6$ \\
$3x + 15y = -1$

```
A = np.array([[14,-2],[3,15]])
B = np.array([6,-1])
# Since this linear algebra problem,
# linalg - linear algebra
C = np.linalg.solve(A,B)
print(C)
```

# Question 17
Plot the polynomial
>$y = 18x^{3}-5x^{2}-2x+3$

over the range $-7\leqslant x \leqslant 1$

```
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-7,1,1000)
y = 18*x**3 - 5*x**2 - 2*x +3
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('y(x)')
```
