# Question 1
Use matrix inverse method to solve the following set,

$
3x_{1} - 4x_{2} + 6x_{3} = 15 \\ 
2x_{1} + 8x_{2} + 12x_{3} = 5 \\
x_{1} - 7x_{2} + 61x_{3} = 10
$

```
import numpy as np
A = np.array([[3,-4,6],[2,8,12],[1,-7,61]])
B = np.array([[15],[5],[10]])
x,res,rank,s = np.linalg.lstsq(A,B,rcond=None)
print(f'x1 = {x[0]}\nx2 = {x[1]}\nx3 = {x[2]}')
```

# Question 2
Use matrix inverse method to solve the following set,

$
3x_{1} - 4x_{2} + 2x_{3} = 5 \\ 
2x_{1} - 3x_{2} + 6x_{3} = 2 \\
6x_{1} - 8x_{2} + 4x_{3} = 3
$

```
A = np.array([[3,-4,2],[2,-3,6],[6,-8,4]])
B = np.array([[5],[2],[3]])
x,res,rank,s = np.linalg.lstsq(A,B,rcond=None)
print(f'rank = {rank}\nx1 = {x[0]}\nx2 = {x[1]}\nx3 = {x[2]}')
# Rank is lower than the number of variables, inverse method is not suitable
```

# Question 3
Represent the following equations in vector-matrix form Ax = b

$
5x - 4y + 3z = 5 \\
4x + 3y - 2z = 23 \\
2x + 6y + 3z = 63
$

The rank of c = [A b] and the value of z are?

```
A = np.array([[5,-4,3],[4,3,-2],[2,6,3]])
B = np.array([[5],[23],[63]])
x,res,c,s = np.linalg.lstsq(A,B,rcond=None)
print(f'rank of c = {c}\nz = {(float(x[2])):.2f}')
```

# Question 4
Use the pseudoinverse method (pinv) to solve the following set of equations

>$
3x_{1} + 15x_{2} + 6x_{3} = 4 \\
x_{1} - 2x_{2} - 3x_{3} = 10
$

$x_{2}$ is equal to:

```
A = np.array([[3,15,6],[1,-2,-3]])
B = np.array([[4],[10]])
A_pinv = np.linalg.pinv(A)
x = A_pinv.dot(B)
print(f'x_2 = {float(x[1])}')
```
