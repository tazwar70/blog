# Question 1
A tank having vertical sides and a bottom area of 100 $ft^{2}$ is used to store
water. The tank is initially empty. To fill the tank, water is pumped into the
top at the rate given in the following table. Determine the water height h(t)
at t = 10 min.


| Time (min) | Flow rate(ft3/min) |
|---|:---|
|0|0|
|1|80|
|2|130|
|3|150|
|4|150|
|5|160|
|6|165|
|7|170|
|8|160|
|9|140|
|10|120|

The relation between height h and the volume inflow rate r is,

>$100\frac{dh}{dt} = r$ &rarr; $h = \frac{1}{100}\int_{0}^{t} r \,dt $

```
import numpy as np
t = np.arange(0,11,1)
r = np.array([0,80,130,150,150,160,165,170,160,140,120])
h = (1/100)*np.trapz(r,t)
print(h)
```

# Question 3
Use the integral function to compute the integral,

$A = \int_{1}^{6} \frac{1}{x} dx$

```
import scipy.integrate as scp
A = lambda x: 1/x
#Quad for regular singular integration
A_int = scp.quad(A,1,6) 
print(A_int[0])
```

# Question 4
Evaluate the following double integral,

$A = \int_{2}^{3}\int_{1}^{4}(x^{2} + xy^{3}) dxdy$

```
A = lambda x,y: (x**2) + (x)*(y**3)
A_int = scp.dblquad(A, 2, 3, lambda x: 1, lambda x: 4)
print(A_int[0])
```

# Question 6
Compute and plot the solution of the following equation,

$10\dot{y} + y = 20 + 7sin(2t)$

$y(0) = 15$.

Find y at t = 60.

```
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define a function which calculates the derivative
def dy_dt(y, t):
    eqn = 20 + 7*np.sin(2*t) - y
    return eqn

ts = np.linspace(0,60,400)
y0 = 15  # the initial condition
ys = odeint(dy_dt, y0, ts)
ys = np.array(ys).flatten()
plt.plot(ys)
plt.xlim([0,70])
```
