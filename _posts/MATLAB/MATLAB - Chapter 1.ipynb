# Question 6
Plot the following functions,
>$y=2\sqrt{x}$ and $z=4sin(3x)$
for $0\leqslant x \leqslant 5$

```
import matplotlib.pyplot as plt
x = np.arange(0,5,0.01)
y = 2*np.sqrt(x)
z = 4*np.sin(3*x)
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('y and z')
plt.legend(['y','z'])
```

# Question 11
The ideal gas law relates the pressure P, volume V,
absolute temperature T, and amount of gas n. The law is
$P = \frac{nRT}{V}$
where R is the gas constant. An engineer must design a large natural gas
storage tank to be expandable to maintain the pressure constant at 2.2 atm.
In December when the temperature is 4Â°F (-15Â°C), the volume of gas in the
tank is 18 500 $ft^{3}$. What will the volume of the same quantity of gas be in
July when the temperature is 88Â°F (31Â°C)? (Hint: Use the fact that n, R, and
P are constant in this problem. Note also that K=C+273.2.)

```
V1 = 18500
T1 = -15+273.2
T2 = 31+273.2
print(f'T1 = {T1}\nT2 = {T2}')
# t1/v1 = t2/v2
# v1/t1 = v2/t2
V2 = (V1*T2)/T1
print(f'The Volume in July = {V2}')
```

# Question 12
The Fourier series is a series representation of a periodic
function in terms of sines and cosines. The Fourier series representation of the
function

> $f(x) = 1 $ ; $ 0 < x < \pi$
> $f(x) = -1 $  ; $-\pi < x < 0$

is
> $\frac{4}{\pi} ( \frac{\sin x}{1} + \frac{\sin 3x}{3} + \frac{\sin 5x}{5} + \frac{\sin 7x}{7} + ... )$

Plot on the same graph the function f(x) and its series representation, using the
four terms shown.

```
pi = np.pi
x1 = np.linspace(0,pi,1000)
f1 = np.linspace(1,1,1000)
x2 = np.linspace(-pi,0,1000)
f2 = np.linspace(-1,-1,1000)
x3 = np.linspace(-pi,pi,1000)
f3 = (4/pi)*( (np.sin(x3))/1 + (np.sin(3*x3))/3 + (np.sin(5*x3))/5 + (np.sin(7*x3))/7)
plt.plot(x1,f1)
plt.plot(x2,f2)
plt.plot(x3,f3)
plt.legend(['f1(x1)','f2(x2)','f2(x2)'])
```
